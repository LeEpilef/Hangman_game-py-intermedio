#**Entorno virtual**

##¿Qué es un entorno virtual?
Un entorno virtual es un ***directorio*** que contiene una instalación de Python de una versión en particular, además de unos cuantos paquetes (módulos) adicionales.

**Inicializar un entrono virtual**
```
python3 -m venv nombre_entorno
```
Se acostumbra llamar al entorno virtual venv o env.

dentro del directorio se creará otro directorio llamado con el nombre_entorno que contiene (bin, include, lib, lib64, pyvenv.cfg y share)

**Activar el entorno virtual**
```
source venv/bin/activate
```
Para activar el entorno virtual mediante un alias se puede usar el siguiente comando:
```
alias avenv="source venv/bin/activate"
```
En el caso de Linux para que alias del comando sea permanente se debe agregar el alias a la configuración de bash o zsh. `sudo nano ~/.zshrc` y luego recargar la terminal con `source ~/.zshrc`.

Usar `deactivate` para **salir** del entorno virtual.

Se recomienda agregar en el archivo .gitignore `venv/` para que el directorio del entorno virtual no quede en el repositorio remoto.

##List comprehensions

Son estructuras que se utilizan para generar nuevas listas.

*Definición:*


```
[element for element in iterable if condition]
```

```
[elemento por elemento en iterable si condición]
```
**element:** elementos a agregar a la lista.

**for element in iterable:** ciclo que genera o extrae los elementos que se agregarán a la lista.

**if condition** (opcional): condición opcional para filtrar los elementos del ciclo.

*el iterable puede ser cualquier otro iterable aparte de un rango; una cadena, lista, diccionario. etc.

````python
#Crea una lista de los numeros naturales del 1 al 100 elevados al cuadrado que no son divisibles por 3.
square_natural_numbers = [number**2 for number in range(1,101) if number % 3 != 0]
print(square_natural_numbers)
````

````python
#Crea una lista de todos los multiplos de 4 que a la vez son multiplos de 6 y de 9 hasta los 5 digitos.
ficty_list = [number for number in range (10000) if number % 4 == 0 and number % 6 == 0 and number % 9 == 0]
print(ficty_list)
#El número 36 es el minimo común multiplo de 4 divisible por 6 y 9 al mismo tiempo por lo que con él se obtiene el mismo resultado evaluando una unica condición.
simplified_ficty_list = [number for number in range (10000) if number % 36 == 0]
print(simplified_ficty_list)
````

En lugar de usar `def` para declarar una función, se utiliza `lambda` y la expresión consiste en una única linea de código. Las funciones lambda no se identifican con un nombre.

##High order functions
*Funciones de ordén superior*

Son funciones que reciben como parámetro a otra función. Las siguientes son tres funciones de orden superior importantes y presentes en muchos lenguajes de programación


- ***filter***

Vamos a usar la función *filter* para filtrar de una lista los número pares, buscando obtener como resultado una lista solo de números impares.

Filter recibe **una función**, en este caso, una lambda function y **un iterable** que es, en este caso, una lista. La función filter retorna un iterador y las función list la que convierte ese iterador en una lista con el resultado que espero.

````python
my_list = [1,4,5,6,9,13,19,21]

odd = list(filter(lambda x: x%2!=0, my_list))
print(odd)
````

- ***map***

Vamos a usar la función *map* para elevar al cuadrado todos los números de una lista y, luego, para transformar el tipo de dato de los elementos de una lista.

Map recibe **una función**, en este caso, una lambda function o la keyword de un tipo de dato y **un iterable** que es, en este caso, una lista. La función map eleva y transforma los valores que contienen las listas en el ejemplo.

````python
my_list = [1,2,3,4,5]

squares = list(map(lambda x: x**2, my_list))
print("cuadrados:", squares)

strings = list(map(str, squares))
print("int -> str:", strings)

my_list = ['1', '2', '3', '4', '5']
int_list = list(map(int, my_list))
print("str -> int:", int_list)
````

Concluyo que si bien tanto filter como map reciben una función y un iterable como parámetros: **filter evalúa** si cada elemento del iterable cumple, o no, una condición y **map ejecuta** la función a cada elemento del iterable.

- ***reduce***

Vamos a usar la función *reduce* para multiplicar entre sí los valores de una lista.

Reduce se debe importar de la libreria functools, esta libreria es nativa de python.

Reduce recibe **una función**, en este caso, una lambda function y **un iterable** que es, en este caso, una lista. La función reduce toma en la primera iteración como párametro *a* el primer 2 y como párametro *b* el segundo 2, de ahí en adelante *a* será el resultado de la multiplicación prevía entre *a* y *b* .

````python
from functools import reduce

my_list = [2,2,2,2,2,2,2,2,2]

all_multiplied = reduce(lambda a, b: a * b, my_list)
print(all_multiplied)
````

Este programa imprime la lista de todos los divisores de un número dado. En su manejo de errores tenemos que:
- Si el dato ingresado por el usuario no se puede transformar a un número entero se le pide que ingrese unicamente un entero y se ejecuta de nuevo la función run()

- Si el número ingresado por el usuario es menor o igual a cero se eleva un **raise ValueError** con un mensaje que informa que el número no es valido.

````python
def divisors(num): 
    #Función que retorna una lista con todos los divisores de un número dado.
    try:
        if num <=0:
            raise ValueError("El número ingresado debe ser positivo.")
        divisors = [divisor for divisor in range(1, num+1) if num % divisor == 0]
        return divisors
    except ValueError as ve:
        print(ve)
        return str(num)+" no es un número valido."

def run():
    try:
        num = int(input("Ingresa un número entero: "))
        divisors_list = divisors(num)
        print(*(divisors_list))
    except ValueError:
        print("Por favor, ingresa unicamnete un número entero.")
        print("")
        run()

if __name__ == "__main__":
    run()
````

#**Manejo de archivos**

Existen dos tipos de archivos:
1. **Texto plano:** .txt, .json, .py, .csv, .xml
2. **Binarios:** .png, .mp4, .mp3, .dll

En programación se acostumbra a manipular casi exclusivamente los archivos de texto plano. 

En python existen diversos modos de apertura para manejar los archivos:

- **r** = Solo lectura
- **r+** = Lectura y escritura
- **w** = Solo escritura. Sobre escribe el archivo si existe. Crea el archivo si no existe
- **w+** = Escritura y lectura. Sobre escribe el archivo si existe. Crea el archivo si no existe
- **a** = Añadido (agregar contenido). Crea el archivo si éste no existe
- **a+** = Añadido (agregar contenido) y lectura. Crea el archivo si éste no existe.

La linea para manejar archivos en python es esta:

```
with open('./ruta_archivo/archivo.ext', 'modo_de_apertura') as variable
```
***with*** = palabra clave, *manejador contextual* que controla el flujo del archivo haciendo que si se cierra el programa o el script finaliza inesperadamente el archivo no se rompa.

La ruta del archivo y el modo de apertura debe estár en formato string.

***as*** = asigna una variable al archivo abierto para referirse a él mediante dicha variable.

el atributo **encoding="utf-8"** como tercer parametro de la linea para manejar archivos evita problemas de codificación.

````python
def create_numbers(file):
  #Con el archivo numbers.txt recien creado, se genera una lista de números del 1 al 100
  for i in range(100):
    file.write(str(i+1)+"\n")


def read():
  #Se lee la lista de números que existe en el archivo y se guarda en una lista.
  numbers = []
  with open("numbers.txt", 'r', encoding="utf-8") as nm:
    for line in nm:
      numbers.append(int(line))
  print(numbers)


def write():
  #Se crea un nuevo archivo llamado names.txt y se escribe en él una lista de nombres.
  names = ["Albert", "Jairo", "Laura", "Miguél", "Marcela"]
  with open("names.txt", 'w', encoding="utf-8") as nm:
    for name in names:
      nm.write(name+"\n")


def run():
  with open('numbers.txt', 'w+') as nm:
    create_numbers(nm)
  nm.close()
  
  read()

  write()

run()
````
